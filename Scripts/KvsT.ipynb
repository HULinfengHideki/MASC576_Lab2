```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = "wrapup.xlsx"
sheet_name = "main"
df = pd.read_excel(file_path, sheet_name=sheet_name)

for target_ns in [1, 2, 3, 4]:
    df_t = df[df["Time(ns)"] == target_ns].copy()

    plt.figure(figsize=(7,5))
    for size in sorted(df_t["Size"].unique()):
        sub = df_t[df_t["Size"] == size].sort_values("Temp")
        plt.plot(
            sub["Temp"],
            sub["K_last4avg (W/m·K)"],
            marker="o",
            linewidth=1.8,
            label=f"L = {size} Å"
        )

    plt.xlabel("Temperature (K)")
    plt.ylabel("Thermal conductivity, K (W/m·K)")
    plt.title(f"K vs T at {target_ns} ns")
    plt.legend()
    plt.tight_layout()
    plt.savefig(f"K_vs_T_{target_ns}ns.png", dpi=300)
    #plt.close()

print("Saved: K_vs_T_1ns.png, K_vs_T_2ns.png, K_vs_T_3ns.png, K_vs_T_4ns.png")
```
